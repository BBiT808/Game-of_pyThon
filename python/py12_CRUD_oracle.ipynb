```python
import oracledb
    # 데이터베이스 연결 시도 !!
try:
    conn = oracledb.connect(
        user="pknu",
        password="1234",
        dsn="oracle/XEPDB1",
    )
    print("Oracle 데이터베이스에 성공적으로 연결 되었습니다 !")
    cursor = conn.cursor()
except Exception as e:
    print("Oracle 데이터베이스 연결 실패: ", e)
```

```python
# 테이블 생성
query = """
    BEGIN
        EXECUTE IMMEDIATE '
            CREATE TABLE students (
            id NUMBER GENERATED BY DEFAULT AS IDENTITY PRIMARY KEY,
            name VARCHAR2(50),
            age NUMBER,
            email VARCHAR2(100)
            )';
    EXCEPTION
        WHEN OTHERS THEN
            IF SQLCODE != -955 THEN
                RAISE;
            END IF;
        END;
"""
cursor.execute(query)
print("테이블 생성 완료 !!")
```

```python
# Create 한 개의 데이터 삽입
que = "INSERT INTO students(name,age,email) VALUES (:1, :2, :3)"
values = ("홍길동",20, "hong@where.myfather")
cursor.execute(que,values)
conn.commit()
print("데이터 삽입이 완료되었습니다! ^_^")
```

```python
# Create 여러 개의 데이터 삽입
que = "INSERT INTO students(name,age,email) VALUES (:1, :2, :3)"
m_values =[("superman",33,"Busan@bb.co.kr"),("Cutegirl",17,"kkk@jeju.co.kr"),("Coolboy",22,"hhh@Jeju.ogg")]
cursor.executemany(que, m_values)
conn.commit()
print("여러 개의 데이터 삽입도 완료 !")
```

```python
# Read 데이터 조회
que = "SELECT * FROM students"
cursor.execute(que)
# 모든 데이터 조회
users = cursor.fetchall()
print("학생목록 : ")
for u in users:
    print(u)
```

```python
# Update 데이터 수정
que = "UPDATE students SET age = :1 WHERE name = :2"
values = (25,"홍길동")
cursor.execute(que,values)
conn.commit()
print("데이터가 성공적으로 수정되었습니다 !")
```
