```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x='day',y='tip', hue = "sex", data = tips, alpha = 0.7, ax=ax1)
sns.swarmplot(x='day',y='tip',hue='sex',data=tips,palette = 'Set2',alpha=0.7,ax=ax2)
ax1.set_title('Strip plot으로 표현한 날짜별로 분류한 성별과 팁 관계')
ax2.set_title('Swarm plot으로 표현한 날짜별로 분류한 성별과 팁 관계')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)


sns.regplot(x="total_bill", y="tip",data = tips, color="hotpink", scatter_kws={"s":50,"alpha":0.5},line_kws={"linestyle":"--","color":"lightgreen"},ax=ax1)
# 한번에 많은 요소를 넣기 위해 딕셔너리형을 채택함 !!! 
sns.regplot(x="total_bill", y="tip",data = tips, color="hotpink", scatter_kws={"s":50,"alpha":0.5},fit_reg=False, ax=ax2)

fig.suptitle("Scatter plots 선형회귀의 ON / OFF 비교 차트")  # supertitle : 제일 위에 공통된 제목 !! 
ax1.set_title('선형회귀 ON')
ax2.set_title('선형회귀 OFF')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue="species", markers=["o","s","D"])
plt.show()
```

```python
fig = plt.figure(figsize=(12,3))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)



sns.boxplot(x="species",y="sepal_length",data=iris, ax=ax1)
sns.violinplot(x="species",y="petal_length",data=iris,hue="species",palette="muted", ax=ax2)
sns.histplot(data=iris,x="petal_length", hue="species", kde=True, element="step",palette="Set2",ax=ax3)

fig.suptitle("붓꽃의 그래프")  # supertitle : 제일 위에 공통된 제목 !! 
ax1.set_title("꽃받침 길이와 종의 관계")
ax2.set_title("종에 다른 꽃잎의 길이 분포 1")
ax3.set_title("종에 다른 꽃잎의 길이 분포 2")


sns.lmplot(x="sepal_length", y="sepal_width",hue="species",data=iris, markers=["o","s","D"], height=6, aspect=1.2)
plt.title("꽃받침 길이와 너비의 관계")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
exercise = sns.load_dataset('exercise')
exercise.sample(5)
```

```python
# point plot
plt.figure(figsize=(10,6))
sns.pointplot(x="time",y="pulse",hue="kind",data=exercise,capsize=0.1)  # capsize는 선 끝에 모자 씌우는 것 
plt.title("심박수와 시간별 운동종류와의 관계")
plt.show()
```

```python
# box plot
plt.figure(figsize=(7,4))
sns.boxplot(x="time",y="pulse",hue="kind",data=exercise,palette="Set2")
plt.title("심박수와 시간별 운동종류와의 관계")
plt.show()
```

```python
# swarm plot
plt.figure(figsize=(7,4))
sns.swarmplot(x="time",y="pulse",hue="kind",data=exercise, dodge=True, palette="Set1")
plt.title("시간과 운동 유형에 따른 개별 맥박 측정")
plt.show()
```

- 결론
> 고강도 운동(Running) : 시간이 지날 수록 심박수가 급격히 상승하며 변동성 또한 상승
> 
> 저강도 운동(Walking) : 심박수는 서서히 증가하지만 변동성은 크지 않음
>
> 휴식(Rest) : 심박수는 거의 일정하게 유지되며 안정적임

> 실무적 적용 : 단시간에 큰 운동 효과를 느끼고 싶다면 고강도 운동을, 자신만의 페이스에 맞게 천천히 운동하고 싶다면 저강도 운동을 선택하는 것이 좋으며, 고강도 운동 후에는 반드시 휴식을 통해 심박수를 안정시켜주는 것이 필요하다.
