```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import datetime

import pandas as pd
import numpy as np
import seaborn as sns

from hdfs import InsecureClient
```

```python
from matplotlib import rc
plt.rc('font',family ="NanumBarunGothic")
plt.text(0.2,0.4,'한글',size=100)
```

```python
hdfs = InsecureClient("http://namenode:9870", user = "hadoop")
paths = ["200812_201312_연령별인구현황_연간.csv","201412_201912_연령별인구현황_연간.csv","202012_202412_연령별인구현황_연간.csv"]
datas = []
for path in paths:
    with hdfs.read(path, encoding="EUC-KR") as reader:
        datas.append(pd.read_csv(reader))

mdata = pd.merge(datas[0], datas[1], on ='행정구역', how = 'left')
data = pd.merge(mdata,datas[2], on='행정구역', how = 'left')
data
```

```python
data = data[data.columns[data.columns.str.contains("행정구역|년_남_|년_여_")]]
data = data[data.columns[~data.columns.str.contains("_총인구수|_연령구간인구수")]] # 물결 ~ 표시는 반대값!! ; 표시 하지 마 !!!
data
```

```python
## pivot과 melt의 차이 ??? ; melt는 긴 데이터 pivot은 넓은 데이터 !!!!!!!!!!
```

```python
data = data.melt(id_vars=['행정구역'],  var_name = '타입', value_name = 'value')
data
```

```python
data['행정구역'] = data['행정구역'].str.replace(r'\(\d+\)','',regex=True) # regex는 정규식을 쓰겠다는 의미래 !!!! ^0^
data    # 이스케이프 문자(\)를 사용해서 (d(숫자))를 다 날려줬음 !!!! ^0^ 대박 신기 근데 외울필요는 없대(지피티가 있어서)
```

```python
test1= data['타입'].str.split('_')
print(type(test1),test1)
test2= data['타입'].str.split('_',expand=True)
print(type(test2),test2)
data[['time','gender','age']] = data['타입'].str.split('_',expand=True)
#data = data[['행정구역','time','gender','age']]
del data['타입']
data.head(5)
```

```python
new_data=data.copy()
new_data.info()
```

```python
new_data['value'].dtype
```

```python
new_data['value'] = new_data['value'].str.replace(',','').fillna(0).astype(int)
new_data.head(3)
```

```python
print(new_data['value'].dtype, new_data.info())
```

```python
new_data['time'] = pd.to_datetime(new_data['time'], format="%Y년")
new_data.head(3)
```

```python
new_data = 
```

```python
# new_data = new_data.drop(index=range(1))
new_data=new_data.drop(new_data.iloc[0:1].index)  # 전국~(노래자랑~) 날림 !!
new_data
```

```python
mydata = new_data.copy()
mydata
```

```python
print(mydata['time'].dtype)
```

```python
# mfilter = mydata['time'] == '2008-01-01'
mfilter = mydata['time'].dt.year == 2008 # 년도만 비교해!! -0-
mydata =  mydata[mfilter]
mydata.sample(5)
```

```python
mydata2 = mydata.groupby(['행정구역','age'])['value'].sum().reset_index()
mydata2.head(15)
```

```python
# mydata2['per'] = mydata2.groupby(['행정구역'])['value'].transform(lambda x : 100 * x / float(x.sum()))
mydata2['per'] = mydata2.groupby(['행정구역'])['value'].apply(lambda x : 100 * x / float(x.sum())).reset_index(level=0,drop=True)
mydata2
```

```python
mydata2 = mydata2.dropna()
mydata2.sample(5)
```

```python
#mydata2 = mydata2.pivot(index="행정구역", columns = "age", values= 'value')
piv_data = mydata2.pivot(index="행정구역", columns = "age", values= 'per')
piv_data
```

```python
print(piv_data.columns)
templist =list(piv_data.columns)
templist.append(templist.pop(1)) # pop으로 빼서 뒤에다 붙이기 !!! ㅇ0ㅇ 짱이다 ~
print(templist)
piv_data = piv_data[templist]
piv_data
```

```python
# 2008년도 지역별 연령 분포 비교
plt.figure(figsize=(8,8))
plt.title("2008년도 지역별 연령 분포 비교")
sns.heatmap(piv_data,cmap="Purples")
plt.savefig("2008년도지역별연령분포비교.png")
plt.show()
```

```python
mydata2 = new_data.copy()
m2filter = mydata2['time'].dt.year == 2024 # 년도만 비교해!! -0-
mydata2 =  mydata2[m2filter]
mydata2.sample(5)
```

```python
mydata2 = mydata2.groupby(['행정구역','age'])['value'].sum().reset_index()
mydata2.head(15)
```

```python
# mydata2['per'] = mydata2.groupby(['행정구역'])['value'].transform(lambda x : 100 * x / float(x.sum()))
mydata2['per'] = mydata2.groupby(['행정구역'])['value'].apply(lambda x : 100 * x / float(x.sum())).reset_index(level=0,drop=True)
mydata2
```

```python
mydata2 = mydata2.dropna()
mydata2.sample(5)
```

```python
#mydata2 = mydata2.pivot(index="행정구역", columns = "age", values= 'value')
piv_data2 = mydata2.pivot(index="행정구역", columns = "age", values= 'per')
piv_data2
```

```python
print(piv_data2.columns)
templist2 =list(piv_data2.columns)
templist2.append(templist2.pop(1)) # pop으로 빼서 뒤에다 붙이기 !!! ㅇ0ㅇ 짱이다 ~
print(templist2)
piv_data2 = piv_data2[templist2]
piv_data2
```

```python
# 2024년도 지역별 연령 분포 비교
plt.figure(figsize=(8,8))
plt.title("2024년도 지역별 연령 분포 비교")
sns.heatmap(piv_data2,cmap="Blues")
plt.savefig("2024년도지역별연령분포비교.png")
plt.show()
```
