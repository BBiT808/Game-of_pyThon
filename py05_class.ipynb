```python
# í…ŒìŠ¤íŠ¸ 9 ì€í–‰ì˜ˆê¸ˆ ìº¡ìŠí™” ì˜ˆì œ, í´ë˜ìŠ¤ ë³€ìˆ˜ ; ìº¡ìŠí™” ë¨ !
class BankAccount:   # ë‹¨ì–´ê°€ ì—°ê²°ì´ ë˜ë©´ ë§ˆë””ì— ëŒ€ë¬¸ìë¥¼ ì“°ê±°ë‚˜, í•˜ì´í”ˆ(-)ìœ¼ë¡œ ì—°ê²°í•˜ê¸°ë„ í•œë‹¤ !
    interest_rate = 0.02     # í´ë˜ìŠ¤ ë³€ìˆ˜ ; ì „ì²´ í•¨ìˆ˜ì— ë‹¤ ì ìš©ì´ ëœë‹¤ëŠ” ëœ» !
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance or 0   # ìº¡ìŠí™” !!!   ; ë„¤ì„ë§¹ê¸€ë§(Name Mangling) ~~
    def bprint(self):
        return f"{self.owner}ë‹˜ì˜ ì”ê³ ëŠ” {self.__balance}ì› ì…ë‹ˆë‹¤."
    def deposit(self, amount):
        self.__balance += amount     # í•œë²ˆ ì–¸ë”ë°”ë¥¼ ë¶™ì´ë©´ ë°‘ì—ë„ ê³„ì† ë¶™ì—¬ì¤˜ì•¼ í•œë‹¤ !
        return f"{self.owner}ë‹˜ì˜ ê³„ì¢Œì— {amount}ì›ì´ ì…ê¸ˆë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡ : {self.__balance}"
    def withdraw(self, amount):
        if self.__balance >= amount :
            self.__balance -= amount
            return f"{self.owner}ë‹˜ì˜ ê³„ì¢Œì— {amount}ì›ì´ ì¶œê¸ˆë˜ì—ˆìŠµë‹ˆë‹¤. í˜„ì¬ ì”ì•¡ : {self.__balance}"
        return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤."
    def interest(self):
        total = self.__balance * (1+self.interest_rate)   # selfë¥¼ ë¬´ì¡°ê±´ ë¶™ì—¬ì„œ ì‚¬ìš©í•´ì•¼ í•œë‹¤ !!
        return f"{self.interest_rate}%ì˜ ì´ìê°€ ë°˜ì˜ëœ ê¸ˆì•¡ì€ {total:.0f}ì› ì…ë‹ˆë‹¤."

#ì¸ìŠ¤í„´ìŠ¤í™”
acc3 = BankAccount("ì² ìˆ˜",50000)
acc4 = BankAccount("ì˜í¬",200000)
print(acc3.bprint())
print(acc4.bprint())
print(acc3.interest())
print(acc4.interest())

# í´ë˜ìŠ¤ ë³€ìˆ˜ ë°”ê¾¸ê¸°
BankAccount.interest_rate = 0.03     # í´ë˜ìŠ¤ ë³€ìˆ˜ëŠ” í†µì§¸ë¡œ ë°”ë€Œì–´ë²„ë¦¬ë‹ˆ ì¡°ì‹¬í•´ì•¼í•¨ !!!
print(acc3.interest())
print(acc4.interest())

# ê³µê²© í›„!!
# acc.balance = 9000000
# print(acc3.balance)

# ë‚´ì¥ë©”ì†Œë“œë¡œ ìº¡ìŠí™”ëœ ë³€ìˆ˜ ì ‘ê·¼
# print(acc3.deposit(30000))
# print(acc3.withdraw(5500))
```

```python
# ì˜ˆì œ !
class VendingMachine:
    def __init__(self):
        """ ìë™íŒë§¤ê¸° ì´ˆê¸°í™”"""
        self.balance = 0   # ì‚¬ìš©ìì˜ ì”ì•¡
        self.items = {    # ìƒí’ˆëª©ë¡ (ì´ë¦„ : ê°€ê²©)
            "ì½œë¼" : 1500,
            "ì‚¬ì´ë‹¤" : 1400,
            "ë¬¼" : 1000
        }
    def insert_money(self, amount):
        """ì”ì•¡ ì¶©ì „"""
        self.balance += amount
        return f"ì”ì•¡ì´ {self.balance}ì›ìœ¼ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤."
    def buy_item(self, item_name):
        """ìƒí’ˆ êµ¬ë§¤"""
        if item_name not in self.items:
            return f"{item_name}ì€(ëŠ”) íŒë§¤í•˜ì§€ ì•ŠëŠ” ìƒí’ˆì…ë‹ˆë‹¤."
        price = self.items[item_name]
        if self.balance < price:
            return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤ ! ({item_name} : {price}ì›, í˜„ì¬ ì”ì•¡ {self.balance}ì›)"
        self.balance -= price
        return f"{item_name}ì„(ë¥¼) êµ¬ë§¤í–ˆìŠµë‹ˆë‹¤! ë‚¨ì€ ì”ì•¡ : {self.balance})ì›"
    def refund(self):
        """ì”ì•¡ ë°˜í™˜"""
        amount = self.balance
        self.balance = 0
        return f"ì”ì•¡ {amount}ì›ì´ ë°˜í™˜ë˜ì—ˆìŠµë‹ˆë‹¤."

# ìë™íŒë§¤ê¸° ê°ì²´ ìƒì„±
vm = VendingMachine()
```

```python
# ì”ì•¡ ì¶©ì „
print(vm.insert_money(2000))  # ì”ì•¡ì´ 2000ì›ìœ¼ë¡œ ì¦ê°€í–ˆìŠµë‹ˆë‹¤.
```

```python
import random

class Character:
    def __init__(self, name, hp, attack_p):
        """ìºë¦­í„° ì´ˆê¸°í™”"""
        self.name = name
        self.hp = hp
        self.attack_p = attack_p
    def attack(self, opponent):
        """ìƒëŒ€ë°©ì„ ê³µê²©"""
        damage = random.randint(2, self.attack_p)
        opponent.hp -= damage
        print(f"{self.name}(ì´)ê°€ {opponent.name}ë¥¼ ê³µê²©í•˜ì—¬ {damage} ë°ë¯¸ì§€ë¥¼ ì…í˜”ìŠµë‹ˆë‹¤.")
    def now_status(self):
        """í˜„ì¬ ì²´ë ¥ ìƒíƒœ í™•ì¸"""
        return f"{self.name} ì²´ë ¥ : {self.hp}"
    def is_alive(self):
        """ìºë¦­í„° ìƒì¡´ ì—¬ë¶€ í™•ì¸"""
        return self.hp > 0
```
